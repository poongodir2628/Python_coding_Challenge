Create a set of hobbies and allow the user to input at least 3 hobbies.


```python
hobbie1=input("Enter the hoppie1:")
hobbie2=input("Enter the hoppie2:")
hobbie3=input("Enter the hoppie3:")
hobbies={hobbie1,hobbie2,hobbie3}
print(hobbies)
```
